```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
df=pd.read_csv(r"C:\Users\anupr\OneDrive\Desktop\laptop_price_prediction\Data\CleanedData.csv")
```

```python
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df
```

```python
y= df['Price_euros']
```

```python
x= df.drop(["Price_euros"], axis=1)
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# Create the imputer (strategy can be 'mean', 'median', or 'most_frequent')
imputer = SimpleImputer(strategy='median')

# Fit on training data and transform it
x_train_imputed = imputer.fit_transform(x_train)

# Don't forget to transform your test data later with the SAME imputer!
# 

# Now train your model
```

```python
x_test_imputed = imputer.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

# 1. Initialize the correct model for continuous numbers
lr = LinearRegression()

# 2. Fit the model on your imputed data
lr.fit(x_train_imputed, y_train)

# 3. Now you can make continuous price predictions!
# predictions = lr.predict(x_test_imputed)
```

```python
import joblib as jb
jb.dump(lr,"linear_regression.lb")
```

```python
lr.score(x_train_imputed,y_train)
```

```python
lr.score(x_test_imputed,y_test)
```

```python
sv =SVR()
sv.fit(x_train_imputed,y_train)
```

```python
import joblib as jb
jb.dump(sv,"support_vector.lb")
```

```python
sv.score(x_train_imputed, y_train)
```

```python
sv.score(x_test_imputed,y_test)
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
jb.dump(dt, "decision_tree.lb")
```

```python
dt.score(x_train, y_train)
```

```python
dt.score(x_test,y_test)
```

```python
rf= RandomForestRegressor()
rf.fit(x_train, y_train)
```

```python
import joblib as jb
jb.dump(rf, "Random_Forest.lb")
```

```python
rf.score(x_train, y_train)
```
